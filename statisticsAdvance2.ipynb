Question 1: What is hypothesis testing in statistics? 

ans: Hypothesis testing is a formal process that evaluates two competing statements (hypotheses) about a population:
- Null Hypothesis (H₀): The default assumption — usually that there is no effect or no difference.

Example: “The average website traffic is 50 visitors per day.”

- Alternative Hypothesis (H₁ or Ha): The statement you want to test — that there is an effect or difference.

Example: “The average website traffic is not 50 visitors per day".

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis? 

ans: The null hypothesis is the default assumption or claim that there is no effect, no difference, or no relationship between variables.
- It represents the status quo or a baseline to test against.
- It is assumed to be true unless evidence suggests otherwise.

Example:
H₀: The average test score of students is equal to 70.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test. 

ans: The significance level, denoted by α (alpha), is the threshold probability used to decide whether the observed data is statistically significant.
- It represents the maximum probability of making a Type I error — rejecting the null hypothesis when it is actually true.
- Common values:
- α = 0.05 (5%) → most commonly used
- α = 0.01 (1%) → stricter
- α = 0.10 (10%) → more lenient


Question 4: What are Type I and Type II errors? Give examples of each.

ans:  Type I Error (False Positive)
- Definition: Rejecting the null hypothesis (H₀) when it is actually true.
- You think there's an effect or difference — but there isn't.
- Controlled by: the significance level (α), typically 0.05.

 Example:
Suppose a medical test is designed to detect a disease.
- H₀: The patient does not have the disease.
- H₁: The patient has the disease.

Type II Error (False Negative)
- Definition: Failing to reject the null hypothesis (H₀) when it is actually false.
- You miss a real effect or difference.
- Controlled by: the power of the test (1 − β).

 Example:
Using the same medical test:
- H₀: The patient does not have the disease.
- H₁: The patient has the disease.
If the test result is negative but the patient actually has the disease, that’s a Type II error.


Question 5: What is the difference between a Z-test and a T-test? Explain when to use each. 

ans:  Z-Test
-  Used when the sample size is large (typically n ≥ 30)
-  Population standard deviation (σ) is known
-  Assumes data follows a normal distribution
-  Uses the standard normal (Z) distribution
-  More precise when population parameters are known

 Commonly used in:
- Comparing sample mean to population mean
- Proportion tests (e.g., A/B testing)
- Quality control and large-scale surveys

 T-Test
-  Used when the sample size is small (typically n < 30)
-  Population standard deviation is unknown
-  Assumes data is approximately normally distributed
-  Uses the Student’s t-distribution
-  Accounts for extra uncertainty due to estimating standard deviation from the sample

 Commonly used in:
- One-sample mean tests
- Two-sample mean comparisons
- Paired sample tests (e.g., before-and-after studies)

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. 

# answer:
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
import numpy as np
from scipy import stats

# Calculate the sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

# Define the null and alternative hypotheses
# H0: mu = 50
# H1: mu != 50

# Perform the Z-test
z_statistic = (sample_mean - 50) / (sample_std / np.sqrt(len(sample_data)))
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

# Display the results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Z-Statistic:", z_statistic)
print("P-Value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. 

# answer:
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate data from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=1000)

# Calculate the 95% confidence interval for the mean
confidence_level = 0.95
degrees_freedom = len(data) - 1
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, loc=sample_mean, scale=sample_std/np.sqrt(len(data)))

# Display the results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("95% Confidence Interval:", confidence_interval)

# Plot the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')
plt.axvline(sample_mean, color='k', linestyle='dashed', linewidth=1)
plt.axvline(confidence_interval[0], color='r', linestyle='dashed', linewidth=1)
plt.axvline(confidence_interval[1], color='r', linestyle='dashed', linewidth=1)
plt.title('Histogram of Simulated Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


# answer:

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = [(x - mean) / std for x in data] if std else [0 for x in data]
    return z_scores

# Generate sample data
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=1000)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.hist(z_scores, bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogram of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Density')
plt.grid()
plt.show()

# Explanation
print("Z-scores represent the number of standard deviations a data point is from the mean.")
```
